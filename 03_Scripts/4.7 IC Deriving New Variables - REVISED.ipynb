#### When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”¶

```python
df_ords_prods_merge.head()
```

```python
orders_per_hour = df_ords_prods_merge ['order_hour_of_day'].value_counts()
```

```python
print(orders_per_hour)
```

```python
top_threshold = orders_per_hour.quantile(0.75)
bottom_threshold = orders_per_hour.quantile(0.25)
```

```python
busiest_period = []

for hour in df_ords_prods_merge["order_hour_of_day"]:
    orders_count = orders_per_hour[hour]  # get total orders for this hour
    if orders_count >= top_threshold:
        busiest_period.append("Most orders")
    elif orders_count <= bottom_threshold:
        busiest_period.append("Fewest orders")
    else:
        busiest_period.append("Average orders")
```

```python
df_ords_prods_merge["busiest_period_of_day"] = busiest_period
```

```python
busiest_period
```

```python
df_ords_prods_merge["busiest_period_of_day"].value_counts()
```

I created a new column, busiest_period_of_day, to classify hours of the day based on order volume. Initially, I made an indexed frequency table to pick the top 5 and bottom 5 hours, but this method seemed too subjective and potentially biased. After conducting research, I remembered the quartile concept from Excel and looked for a similar solution in Python. Rather than picking the busiest and slowest hours manually, I decided to use percentiles to make the classification more systematic and data-driven. The top 25% of hours by order count were labeled "Most orders", the bottom 25% were labeled "Fewest orders", and the remaining hours were labeled "Average orders". This approach ensures that the labeling adapts to the actual distribution of orders, avoids arbitrary decisions, and provides a clear overview of peak, average, and low-activity periods. I also checked the resulting counts for each category to confirm the labeling was applied correctly.¶
